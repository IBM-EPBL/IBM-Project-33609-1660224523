## Dropping Unnecessary columns

```python
df = df.drop('Unnamed: 25',axis=1)
df.isnull().sum()
```

```python
df
```

```python
df= df[["FL_NUM", "MONTH", "DAY_OF_MONTH", "DAY_OF_WEEK", "ORIGIN", "DEST", "CRS_ARR_TIME","DEP_DEL15", "ARR_DEL15"]]
df.isnull().sum()
```

```python
df = df.fillna ({'ARR_DEL15': 1})
df = df.fillna ({'DEP_DEL15': 0})
df.iloc[177:185]
```

```python
import math
for index, row in df.iterrows():
    df.loc[index, 'CRS_ARR_TIME'] = math.floor(row['CRS_ARR_TIME'] / 100)
df.head()
```

## Label encoding and hot encoding

```python
#le = LabelEncoder()
#df['DEST'] = le.fit_transform(df [ 'DEST'])
#df['ORIGIN'] =le.fit_transform(df ['ORIGIN'])
```

```python
df
```

## Splitting the dataset into dependent and independent variables

```python
df =pd.get_dummies (df, columns=['ORIGIN', 'DEST']) 
df.head()
```

```python
df=df[["FL_NUM","MONTH","DAY_OF_MONTH","DAY_OF_WEEK","CRS_ARR_TIME","DEP_DEL15","ORIGIN_ATL","ORIGIN_DTW","ORIGIN_JFK","ORIGIN_MSP","ORIGIN_SEA","DEST_ATL","DEST_DTW","DEST_JFK","DEST_MSP","DEST_SEA","ARR_DEL15"]]
df
```

```python
x=df.iloc[:, 0:16].values
y=df.iloc[:, 16:17].values
```

```python
x
```

```python
y
```

## Split the dataset into train set and test set

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.2, random_state=0)
```

```python
x_train
```

```python
x_test.shape
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
y_train.shape
```

## Encoding Categorical cloumns to numerical

```python
sc= StandardScaler()
x_train= sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

## Balancing Dataset Using SMOTE Technique

```python
smote=SMOTE()
x_train_smote,y_train_smote =smote.fit_resample(x_train,y_train)
```

# Model Building

## DecisionTree 

```python
dmodel=DecisionTreeClassifier(random_state=100)
```

```python
dmodel.fit(x_train_smote,y_train_smote)
```

```python
ypredd=dmodel.predict(x_test)
```

## Random Forest 

```python
from sklearn.ensemble import RandomForestClassifier
Rmodel=RandomForestClassifier(n_estimators=100)
```

```python
Rmodel.fit(x_train_smote,y_train_smote)
```

```python
ypredR=Rmodel.predict(x_test)
```

# Compare The Model

```python
print("Decision Tree Model Testing Accuracy")
print(accuracy_score(y_test,ypredd))
```

```python
print("Random Forest Model Testing Accuracy")
print(accuracy_score(y_test,ypredR))
```

# Evaluating Performance Of The Model And Saving The Model

### Decision Tree is Selected 

```python
f1_score(ypredd,y_test,average='weighted')
```

```python
print(confusion_matrix(y_test,ypredd))
```

```python
print(classification_report(y_test,ypredd))
```
